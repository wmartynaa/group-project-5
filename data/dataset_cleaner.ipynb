```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

df = pd.read_csv('globalterrorismdb_0522dist.csv', sep=';')
# Delete columns that represent textual information in numbers
df = df.drop(columns = ['country', 'region', 'alternative', 'attacktype1', 'attacktype2', 'attacktype3', 'targtype1', 'targsubtype1', 'corp1', 'target1', 'natlty1',  'targtype2', 'targsubtype2', 'natlty2', 'targtype3', 'targsubtype3', 'natlty3', 'claimmode', 'claimmode2', 'claimmode3', 'weaptype1',  'weapsubtype1', 'weaptype2', 'weapsubtype2', 'weaptype3', 'weapsubtype3',  'weaptype4', 'weapsubtype4', 'propextent', 'propvalue', 'propcomment', 'hostkidoutcome'])

df.head()
```

```python
#U5 perpetrator information
def map_perpetrator_type(perpetrator_type):
    if perpetrator_type != 'Unknown':
        return 10*0.0279
    else:
        return 6*0.0279
    
df['perpetrator_type_value'] = df['gname'].apply(map_perpetrator_type)
```

```python
#U5 value for attack mode
def map_attack_type(attack_type):
    if attack_type in ['Assassination', 'Armed Assault', 'Bombing/Explosion', 'Hostage Taking (Kidnapping)']:
        return 10*0.0177
    elif attack_type in ['Hijacking', 'Facility/Infrastructure Attack', 'Hostage Taking (Barricade Incident)']:
        return 7*0.0177
    else:
        return 3*0.0177

df['attack_type_value'] = df['attacktype1_txt'].apply(map_attack_type)

```

```python
#U5 value for weapon type
def map_weapon_type(weapon_type):
    if weapon_type in ['Radiological', 'Biological', 'Chemical']:
        return 10*0.0094
    else:
        return 7*0.0094
    
df['weapon_type_value'] = df['weaptype1_txt'].apply(map_weapon_type)
```

```python
#U4 value for target type
def map_target_type(target_type):
    if target_type in ['Government (General)', 'Police', 'Military', 'Government (Diplomatic)']:
        return 10*0.0455
    elif target_type in ['Food or Water Supply', 'Transportation', 'Telecommunication', 'Educational Institution', 'Airports & Aircraft']:
        return 7*0.0348
    else:
        return 4*0.0301

df['target_type_value'] = df['targtype1_txt'].apply(map_target_type)
```
