```python
import pandas as pd
import os

# Change the current directory to the new directory
os.chdir('D:/Cssci/terrorism/group-project-5')

# load dataset
df = pd.read_excel('data/globalterrorismdb_0522dist.xlsx')
```

```python
display(df)
```

```python
df = df[['attacktype1_txt','weaptype1_txt','multiple','success','doubtterr','nkill','nwound','nhostkid']]
```

```python
display(df)
```

```python
df["severity"] = df['nkill'] + df ["nwound"]
df = pd.get_dummies(df, columns=['attacktype1_txt', 'weaptype1_txt'])
df.drop(columns=['attacktype1_txt_Unknown', 'weaptype1_txt_Unknown'], inplace=True)
```

```python
display(df)
```

```python
# Fill missing values with zero
df.fillna(0, inplace=True)

# Convert the DataFrame to integer type
df = df.astype(int)
```

```python
import statsmodels.api as sm

X = df.drop(columns=['nkill','nwound','severity','nhostkid'])
y = df['nhostkid'] 
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```

```python
import sys
with open('model_summary.txt', 'w') as f:
    # Redirect stdout to the file
    sys.stdout = f
    # Print the model summary
    print(model.summary())
    # Reset stdout
    sys.stdout = sys.__stdout__

# Read the saved text file and display its contents
with open('model_summary.txt', 'r') as f:
    print(f.read())
```

```python
with open('nhostkid.html', 'w') as f:
    f.write(model.summary().as_html())
```
